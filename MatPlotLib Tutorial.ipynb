```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#Resize your graph (dbi specifies pixels per inch, ideally it should be 300)
plt.figure(figsize=(8,5),dpi=100)

#Keyword Argument Notation
plt.plot(x,y,color='red',linewidth=2,marker='.',linestyle='--',markersize=10,markeredgecolor='blue')
```

```python
plt.plot(x,y,'b^--')
```

```python
'''
b:blue
g:green
r:red
c:cyan
m:magenta
y:yellow
k:black
w:white
'''
```

```python
#Select interval we want to plot points at
x2=np.arange(0,4.5,0.5)
print(x2)

#Plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'r',label='X^2')

plt.legend()
```

```python
plt.plot(x2[5:],x2[5:]**2,'r--')

#Add a title(specify font parameters with fontdict)
plt.title('Our First Graph!',fontdict={'fontname':'Comic Sans MS','fontsize':20})
```

```python
plt.plot(x2[5:],x2[5:]**2,'r--',label='X')

#Add a title(specify font parameters with fontdict)
plt.title('Our First Graph!',fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.savefig('mygraph.png',dpi=300)
```

```python
labels=['A','B','C']
values=[1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars=plt.bar(labels,values)

patterns=['/','o','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
gas.USA  # same as gas['USA']
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices Over Time (in USD)',fontdict={'fontweight':'bold','fontsize':18})

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_Price_Figure.png',dpi=300)

plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall,bins=bins,color='c')

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')

plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('Histogram.png',dpi=300)

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left','Right']
colors = ['#abcdef','#aabbcc']

plt.pie([left,right],labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```
